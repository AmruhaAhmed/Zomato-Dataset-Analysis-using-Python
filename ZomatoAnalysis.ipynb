```python
from wordcloud import WordCloud, STOPWORDS
```

```python
cuisine_words=""
stopwords=set(STOPWORDS)
```

```python
for val in df.Cuisines:
    val=str(val)
    tokens=val.split()
    for i in range(len(tokens)):
        tokens[i]=tokens[i].lower()
    cuisine_words+=" ".join(tokens)+" "
```

```python
wordcloud=WordCloud(min_font_size=5).generate(cuisine_words)
```

```python
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Most popular Cuisines across the Globe")
```

```python
locality_words=""
stopwords=set(STOPWORDS)
```

```python
for val in df.Locality:
    val=str(val)
    tokens=val.split()
    for i in range(len(tokens)):
        tokens[i]=tokens[i].lower()
    locality_words+=" ".join(tokens)+" "
```

```python
wordcloud=WordCloud(min_font_size=5,collocations=False).generate(locality_words)
```

```python
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Most popular Localities across the Globe")
```

```python
from pywaffle import Waffle
```

```python
def agg_rating(input):
    if(input<=1):
        return 1
    elif(input>1 and input<=2):
        return 2
    elif(input>2 and input<=3):
        return 3
    elif(input>3 and input<=4):
        return 4
    else:
        return 5
```

```python
df['Aggregate_Rating']=df['Aggregate_Rating'].apply(agg_rating)
```

```python
value=list(df['Aggregate_Rating'].value_counts())
print(value)
```

```python
plt.figure(
    FigureClass = Waffle,
    rows = 50,columns=50,
    values = value,title={"label":"Waffle Chart of Aggregate_Rating"},legend={"labels":["1 star","2 star","3 star","4 star","5 star"]})
```
